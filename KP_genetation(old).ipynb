```python
import openai
import yaml
with open("conf/index.yaml") as f:
    credentials = yaml.safe_load(f)
openai.api_key = credentials['environment_variables']['OPENAI_API_KEY']
model_name = "gpt-4"
system_prompt = ""
user_query = ""
```

```python
import pandas as pd
df = pd.read_csv('./data/ground_truth.csv')
print(df.head(10))
# df = df.head(10)
arguments = list(df['argument'].values)
target_KPs = list(df['key_point'].values)
print(arguments)
print(len(arguments))
```

```python
predict_KPs = []
system_prompt = \
        """
        You need to do key point analysis on the query of user and
        generate a key point from it.
        """
for argument in arguments:
    rep = openai.ChatCompletion.create(
        model = model_name,
        messages = [
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": argument}
        ]
    )

    result = rep["choices"][0]["message"]["content"]
    print(len(predict_KPs))
    print(result)
    predict_KPs.append(result)
```

### Read result from csv file

```python
import pandas as pd

# Specify the CSV file path
csv_file_path = 'result/gpt-4.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Extract each column as a separate list
target = df['target_KP'].tolist()
predict = df['predict_KP'].tolist()

# Print or use the lists as needed
print("Predict_KPs:", predict)
print("Target_KPs: ", target)
```

### BLEU Score

```python
import sacrebleu

# Example sentences
candidates = predict
references = target

# Join sentences into strings
candidate_str = "\n".join(candidates)
reference_str = "\n".join(references)

# Calculate BLEU score
bleu = sacrebleu.raw_corpus_bleu(candidate_str, [reference_str])
print("BLEU Score:", bleu.score)
```

### BERT Score

```python
from evaluate import load
bertscore = load("bertscore")
result = bertscore.compute(predictions=predict, references=target, lang="en")['f1']
print(result)
print(sum(result)/len(result))
```

### Rouge Score

```python
from rouge import Rouge
rouge = Rouge()
reference = " ".join(target)
candidate = " ".join(predict)
rouge_scores = rouge.get_scores(candidate, reference)

print("ROUGE Scores:")
print(f"ROUGE-1 Precision: {rouge_scores[0]['rouge-1']['p']}")
print(f"ROUGE-1 Recall: {rouge_scores[0]['rouge-1']['r']}")
print(f"ROUGE-1 F1 Score: {rouge_scores[0]['rouge-1']['f']}")

print(f"ROUGE-2 Precision: {rouge_scores[0]['rouge-2']['p']}")
print(f"ROUGE-2 Recall: {rouge_scores[0]['rouge-2']['r']}")
print(f"ROUGE-2 F1 Score: {rouge_scores[0]['rouge-2']['f']}")

print(f"ROUGE-L Precision: {rouge_scores[0]['rouge-l']['p']}")
print(f"ROUGE-L Recall: {rouge_scores[0]['rouge-l']['r']}")
print(f"ROUGE-L F1 Score: {rouge_scores[0]['rouge-l']['f']}")
```
