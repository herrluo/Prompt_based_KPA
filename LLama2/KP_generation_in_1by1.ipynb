### Predict via LLama2

```python
import yaml
with open("../conf/index.yaml") as f:
    credentials = yaml.safe_load(f)
Llama2_api_token = credentials['environment_variables']['LLAMA2_API_TOKEN']
```

```python
import ast
import re
def generating(topic, argument):
    # print(topic)
    # print(argument)
    import replicate
    replicate = replicate.Client(api_token=Llama2_api_token)
    output = replicate.run(
        "meta/llama-2-70b-chat",
        input={
            "debug": False,
            "top_p": 1,
            "prompt": argument,
            "temperature": 0.5,
            "system_prompt": f"""
        You need to do key point analysis on the query of user and
        generate a key point in one sentence from it based on the topic "{topic}" . You should only return the key point.
        """,
        #     "system_prompt": f"""
        # You need to do key point analysis on the query of user and
        # generate a key point in one sentence from it . You should only return the key point.
        # """,
            "max_new_tokens": 500,
            "min_new_tokens": -1,
            "prompt_template": "[INST] <<SYS>>\n{system_prompt}\n<</SYS>>\n\n{prompt} [/INST]",
            "repetition_penalty": 1.15
        }
    )
    # print(output)
    for item in output:
    # https://replicate.com/meta/llama-2-70b-chat/api#output-schema
        print(item, end="")
    output_sentence = ' '
    for item in output:
        output_sentence += item
    # print(output_sentence)
    # print("-------------------")
    # print("0: "+output_sentence[0])
    # print("1: "+output_sentence[1])
    # print("2: "+output_sentence[2])
    # print("3: "+output_sentence[3])
    # print("-1: "+output_sentence[-1])
    # print("----------------------")
    if ':' in output_sentence:
        colon_index = output_sentence.index(':')
        output_sentence = output_sentence[colon_index+1:]
    print(output_sentence)
    # print("*******************")
    # print("0: "+output_sentence[0])
    # print("1: "+output_sentence[1])
    # print("2: "+output_sentence[2])
    # print("3: "+output_sentence[3])
    # print("-1: "+output_sentence[-1])
    # print("*******************")
    #Find the index of first char
    start_index = 0
    for index,char in enumerate(output_sentence):
        if char!= ' ' and char!= '"' and char!= '\n' :
            start_index = index
            break
    output_sentence = output_sentence[start_index:]
    if output_sentence[-1] == '"':
        output_sentence = output_sentence[:-1]
    return output_sentence
    # if output_sentence[1] == '"' and output_sentence[-1] == '"' :
    # # Remove the first and last characters (double quotes)
    #     modified_text = output_sentence[2:-1]
    #      # print(modified_text)
    #     return modified_text
    # else:
    #     # print("The string does not start and end with double quotes.")
    #     return output_sentence
    # The meta/llama-2-70b-chat model can stream output as it's running.
    # The predict method returns an iterator, and you can iterate over that output.
    # if not any([':' in output,'\n\n*' in output]):
    #     print("Neither ':' nor '\n\n*' found in output. Regenerating string...")
    #     return generating(topic, argument)
    # else:
    #     if '\n\n*' in output :
    #         for item in output:
    #             output_sentence += item
    #         print(output_sentence)
    #         start_index = output_sentence.find('*')
    #         extracted_sentence = output_sentence[start_index + 2:].strip()
    #         return extracted_sentence
    #     else :
    #         for item in output:
    #             output_sentence += item
    #         print(output_sentence)
    #         start_index = output_sentence.find(':')
    #         extracted_sentence = output_sentence[start_index + 1:].strip()
    #         return extracted_sentence

```

```python
for index, row in df.iloc[6660:6661].iterrows():
    print(str(index)+": ")
    topic = row['topic']
    argument = row['argument']
    print("argument: "+argument)
    result = generating(topic, argument)
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    print(result)
    df.at[index, 'key_point'] = result
```

```python
import re
def contains_chinese(text):
    # Regular expression pattern for Chinese characters
    chinese_pattern = re.compile('[\u4e00-\u9fff]+')

    # Search for Chinese characters in the text
    if chinese_pattern.search(text):
        return True
    else:
        return False
```
