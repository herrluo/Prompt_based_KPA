### Extract all arguments from each stance of a topic

```python
import pandas as pd
df = pd.read_csv('./data/ArgKP-2021_dataset.csv')
ArgKP2021_topics = [
    "Assisted suicide should be a criminal offence",
    "Homeschooling should be banned",
    "The vow of celibacy should be abandoned",
    "We should abandon marriage",
    "We should abandon the use of school uniform",
    "We should abolish capital punishment",
    "We should abolish intellectual property rights",
    "We should abolish the right to keep and bear arms",
    "We should adopt an austerity regime",
    "We should adopt atheism",
    "We should adopt libertarianism",
    "We should ban human cloning",
    "We should ban private military companies",
    "We should ban the use of child actors",
    "We should close Guantanamo Bay detention camp",
    "We should end affirmative action",
    "We should end mandatory retirement",
    "We should fight for the abolition of nuclear weapons",
    "We should fight urbanization",
    "We should introduce compulsory voting",
    "We should legalize cannabis",
    "We should legalize prostitution",
    "We should legalize sex selection",
    "We should prohibit flag burning",
    "We should prohibit women in combat",
    "We should subsidize journalism",
    "We should subsidize space exploration",
    "We should subsidize vocational education",
    "Routine child vaccinations should be mandatory",
    "Social media platforms should be regulated by the government",
    "The USA is a good country to live in"
]
arg_arguments = []
```

```python
def remove_duplicate_sentences(sentences):
    unique_sentences = []
    seen_sentences = set()

    for sentence in sentences:
        if sentence not in seen_sentences:
            unique_sentences.append(sentence)
            seen_sentences.add(sentence)

    return unique_sentences
```

```python
for topic in ArgKP2021_topics:
    topic_p_df = df[(df['topic'] == topic) & (df['stance'] == 1)]
    topic_n_df = df[(df['topic'] == topic) & (df['stance'] == -1)]
    topic_p_arguments = list(topic_p_df['argument'].values)
    topic_n_arguments = list(topic_n_df['argument'].values)
    deduplicated_topic_p = remove_duplicate_sentences(topic_p_arguments)
    deduplicated_topic_n = remove_duplicate_sentences(topic_n_arguments)
    deduplicated_topic_p = '\n'.join(deduplicated_topic_p)
    deduplicated_topic_n = '\n'.join(deduplicated_topic_n)
    arg_arguments.append([deduplicated_topic_p,deduplicated_topic_n])
```

### Generation Part of KPA

```python
import openai
import yaml
with open("conf/index.yaml") as f:
    credentials = yaml.safe_load(f)
openai.api_key = credentials['environment_variables']['OPENAI_API_KEY']
model_name = credentials['environment_variables']['OPENAI_FINETUNED_MODEL_NAME']
system_prompt = ""
user_query = ""
```

```python
import ast
predict_KPs = []
test_arg = arg_arguments[:10]
for topic in arg_arguments:
    result_topic = []
    system_prompt = f"""
        You need to do key point analysis on a set of arguments from user. They are all about the topic "{ArgKP2021_topics[arg_arguments.index(topic)]}" You need to generate 10-12 pieces of key points and stored them in a list and then return this list.
        """
    for instance in topic:
        print(system_prompt)
        rep = openai.ChatCompletion.create(
            model = model_name,
            messages = [
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": instance}
            ]
        )

        result = rep["choices"][0]["message"]["content"]
        result_topic.append(result)
        print(len(predict_KPs))
        print(result)
    predict_KPs.append(result_topic)
```

```python
print(len(predict_KPs))
```

```python
f = open("./result/gpt-4_per_topic_stance.txt",'w', encoding='UTF-8')
for batch_id, kp in enumerate(predict_KPs):
    print(batch_id)
    print("-------------------Processing-------------------")
    f.write(str(batch_id)+"\n")
    f.write("topic:\n")
    f.write(ArgKP2021_topics[batch_id]+"\n")
    f.write("stance:\n")
    f.write("positive:\n")
    f.write(kp[0]+"\n")
    f.write("negative:\n")
    f.write(kp[1]+"\n")
    f.write("\n")

print("-----------------Finished--------------------------")
```

```python
import ast
final_result = []
for topic_result in predict_KPs:
    t_result = []
    for stance_result in topic_result:
        stance_result = stance_result.rstrip('.').rstrip()
        print(type(stance_result))
        print(stance_result)
        result = ast.literal_eval(stance_result)
        t_result.append(result)
    final_result.append(t_result)
```
